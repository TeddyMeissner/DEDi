```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
NumParameters = [40,35,30,25,20,15,10,5]
AICValues = [30,25,20,15,12,10,9,15]

NumParametersAdaptive = [9,8,7,6]
AICValuesAdaptive = [8.5,8,7.8, 13]

# Plotting the first sweep points
plt.scatter(NumParameters, AICValues, c='black', label='First Sweep')
for i, (x, y) in enumerate(zip(NumParameters, AICValues)):
    plt.text(x, y + 0.5, str(i + 1), color='black', fontsize=10, ha = 'center')

# Plotting the adaptive sweep points
plt.scatter(NumParametersAdaptive, AICValuesAdaptive, c='blue', label='Adaptive')
for i, (x, y) in enumerate(zip(NumParametersAdaptive, AICValuesAdaptive), start=len(NumParameters)):
    plt.text(x, y + 0.5, str(i + 1), color='blue', fontsize=10, ha = 'center')

plt.xlabel('Number of Model Terms')
plt.ylabel('BIC Value')
plt.title("Adaptive Model Selection - Forward")
plt.legend()
plt.grid(True, color='lightgrey', linestyle='-', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the first sweep
NumParameters = [40, 35, 30, 25, 20, 15, 10, 5]
AICValues = [30, 25, 20, 15, 12, 10, 9, 15]

# Data for the adaptive sweep
NumParametersAdaptive = [9, 11, 12, 13]
AICValuesAdaptive = [10, 8, 7.8, 9]

# Plotting the first sweep points
plt.scatter(NumParameters, AICValues, c='black', label='First Sweep')
for i, (x, y) in enumerate(zip(NumParameters, AICValues)):
    plt.text(x, y + 0.5, str(i + 1), color='black', fontsize=10, ha = 'center')

# Plotting the adaptive sweep points
plt.scatter(NumParametersAdaptive, AICValuesAdaptive, c='blue', label='Adaptive')
for i, (x, y) in enumerate(zip(NumParametersAdaptive, AICValuesAdaptive), start=len(NumParameters)):
    plt.text(x, y + 0.5, str(i + 1), color='blue', fontsize=10, ha = 'center')

# Setting labels and title
plt.xlabel('Number of Model Terms')
plt.ylabel('BIC Value')
plt.title("Adaptive Model Selection-Backwards")
plt.legend()
plt.grid(True, color='lightgrey', linestyle='-', linewidth=0.5)

plt.show()
```
