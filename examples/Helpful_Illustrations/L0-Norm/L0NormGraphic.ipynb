```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def l0Smooth(theta,eps):
    return 1 - np.e**(-theta**2/eps**2)
```

```python
x = np.linspace(-2,2,1000)

L0true = np.ones_like(x)
plt.plot(x,L0true,label = 'True',c = 'b')
plt.scatter(0,0,c = 'b',marker='o')

eps_string = [1,.5,.1,.01]

for eps in eps_string:
    plt.plot(x,l0Smooth(x,eps),label='$\epsilon$ = ' + str(eps))
    plt.legend()

plt.ylabel('1 - $\exp(-\\theta^2/\epsilon^2)$')
plt.xlabel('$\\theta$')
plt.title('Regularized $l_0$')
plt.savefig('RegularizedL0.pdf', format='pdf', bbox_inches='tight')
```
