```python
import numpy as np
import matplotlib.pyplot as plt
from ODEFuncs import *
```

```python
# Lorenz Data setup
TSpan = (0, 10)
x0 = [-8.0, 8.0, 27.0]
dt = 0.001

Lorenz_Data = {}
# Initialize nested dictionaries for this dt
t, x_data_true = lorenzData(TSpan, dt, x0)
Lorenz_Data['x'] = x_data_true
Lorenz_Data['t'] = t

np.savez("../LorenzData.npz", x=Lorenz_Data['x'], t=Lorenz_Data['t'])
```

```python
TSpan = (0, 10)
x0 = [0.0, 2.0]
dt = 0.001
VDP_Data = {}
t, x_data_true = vanderpolData(TSpan, dt, x0)
VDP_Data['x'] = x_data_true
VDP_Data['t'] = t

np.savez("../VDPData.npz", x=VDP_Data['x'], t=VDP_Data['t'])
```

```python
TSpan = (0, 10)
# Number of variables
N = 5
F = 8.0  # Forcing term, often used to induce chaotic behavior
# Initial condition: small perturbation around F
x0 = F * np.ones(N)
x0[0] += 0.01

dt = 0.001
Lorenz96_Data = {}
t, x_data_true = lorenz96Data(TSpan, dt, x0)
Lorenz96_Data['x'] = x_data_true
Lorenz96_Data['t'] = t

np.savez("../Lorenz96Data.npz", x=Lorenz96_Data['x'], t=Lorenz96_Data['t'])
```

```python
# Initial conditions and parameters
x0 = [0.01, 0.00, 0.00]       # initial voltage v and recovery variable w
TSpan = (0, 50.0)      # integration time span
dt = 0.01
Colpitts_Data = {}
t, x_data_true = colpittsData(TSpan, dt, x0)
Colpitts_Data['x'] = x_data_true
Colpitts_Data['t'] = t

np.savez("../ColpittsData.npz", x=Colpitts_Data['x'], t=Colpitts_Data['t'])
```
